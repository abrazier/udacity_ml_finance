```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

def get_data(symbols, dates):
    df = pd.DataFrame(index=dates)
    if 'SPY' not in symbols:
        symbols.insert(0, 'SPY')

    for symbol in symbols:
        df_temp = pd.read_csv("{}.csv".format(symbol), index_col='Date', parse_dates=True, usecols = ['Date', 'Adj Close'], na_values=['nan'])
        df_temp = df_temp.rename(columns = {'Adj Close' : symbol})
        df = df.join(df_temp)
        if symbol == 'SPY':
            df = df.dropna(subset=["SPY"])

    return df

def normalize_data(df):
    return df / df.ix[0,:]

def plot_data(df):
    ax = df.plot()
    ax.set_title("Stock Prices", fontsize = 12, color = 'white')
    ax.set_xlabel("Date", color = 'white')
    ax.set_ylabel("Price", color = 'white')
    ax.tick_params(axis = 'both', color = 'white', labelcolor = 'white')
    myFmt = mdates.DateFormatter("%m/%y")
    ax.xaxis.set_major_formatter(myFmt)
    plt.show()

def test_run():
    dates = pd.date_range('2010-01-01', '2012-12-31')
    symbols = ['SPY', 'XOM', 'GOOG', 'GLD']
    df = get_data(symbols, dates)
    plot_data(df)
    plot_data(normalize_data(df))
    
    
    #stats
    print("Mean of df:\n", df.mean())
    print("\nMedian of df:\n", df.median())
    print("\nStd Dev of df:\n", df.std())
    
if __name__ == "__main__":
    test_run()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

def get_data(symbols, dates):
    df = pd.DataFrame(index=dates)
    if 'SPY' not in symbols:
        symbols.insert(0, 'SPY')

    for symbol in symbols:
        df_temp = pd.read_csv("{}.csv".format(symbol), index_col='Date', parse_dates=True, usecols = ['Date', 'Adj Close'], na_values=['nan'])
        df_temp = df_temp.rename(columns = {'Adj Close' : symbol})
        df = df.join(df_temp)
        if symbol == 'SPY':
            df = df.dropna(subset=["SPY"])

    return df

def get_rolling_mean(values, window):
    return pd.rolling_mean(values, window=window)
    
def get_rolling_std(values, window):
    return pd.rolling_std(values, window=window)

def get_bollinger_bands(rm, rstd):
    upper_band = rm + rstd * 2
    lower_band = rm - rstd * 2
    return upper_band, lower_band

def test_run():
    dates = pd.date_range('2012-01-01', '2012-12-31')
    symbols = ['SPY']
    df = get_data(symbols, dates)
    
    #calulate rolling mean, rolling std dev, and bollinger bands
    rm_SPY = get_rolling_mean(df['SPY'], window=20)
    rstd_SPY = get_rolling_std(df['SPY'], window=20)
    upper_band, lower_band = get_bollinger_bands(rm_SPY, rstd_SPY)
    
    
    #plot all the series
    ax = df['SPY'].plot(title="Bollinger Bands", label="SPY")
    rm_SPY.plot(label = 'Rolling Mean', ax=ax)
    upper_band.plot(label='Upper Band', ax=ax)
    lower_band.plot(label='Lower Band', ax=ax)
    
    
    
    
    #Graph settings
    ax.set_title("SPY Rolling Mean", fontsize = 12, color = 'white')
    ax.set_xlabel("Date", color = 'white')
    ax.set_ylabel("Price", color = 'white')
    ax.tick_params(axis = 'both', color = 'white', labelcolor = 'white')
    myFmt = mdates.DateFormatter("%m/%y")
    ax.xaxis.set_major_formatter(myFmt)
    ax.legend(loc='upper left')
    plt.show()
        
    
    
if __name__ == "__main__":
    test_run()
```
